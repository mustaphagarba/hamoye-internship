Time to use the following cells to answer questions for the quiz

```python
# Question: Which of the following fuel type code has the lowest average fuel cost per unit burned?
print(data['fuel_cost_per_unit_burned'])
```

```python
# Question: What is the standard deviation and 75th percentile of the measure of energy per unit (Fuel_mmbtu_per_unit) in two decimal places?
data['fuel_mmbtu_per_unit'].describe()
```

```python
# What is the skewness and kurtosis for the fuel quantity burned in two decimal places?

kurt = data['fuel_qty_burned'].kurt()
skew = data['fuel_qty_burned'].skew()
print('Kurtosis is ', kurt)
print('Skew is ', skew)
```

```python
# Question: Which of the features has the second and third lowest correlation with the Fuel Cost Per Unit Burned?
data.corr()
```

```python
# Question: For the fuel type coal, what is the percentage change in the fuel cost per unit burned in 1998 compared to 1994?
data5 = data[data.report_year == 1994]
data6 = data[data.report_year == 1998]
res = data5['fuel_cost_per_unit_burned'].mean()
res2 = data6['fuel_cost_per_unit_burned'].mean()
print(((res - res2)/res2) *100 ) 
```
